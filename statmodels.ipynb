```python
# import necessary libraries
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import sklearn as scikit_learn
```

```python
df = pd.read_csv('model_data.csv', index_col=0)
df.head(3)
```

```python
# Define the dependent Y and independent variables X
X = df[['Population']]
Y = df['Productivity']

# Add constant term to the independent variables
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.OLS(Y, X)
results = model.fit()

# Print the logistic regression results
print(results.summary())
```

```python
# Define the dependent Y and independent variables X
X = df[['Population', 'Gdp', 'Compensation', 'CapitalStock', 'earlyLeavers']]
Y = df['Productivity']

# Add constant term to the independent variables
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.OLS(Y, X)
results = model.fit()

# Print the logistic regression results
print(results.summary())

# Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF results
print(vif_data)
```

VIF Calculation: For each independent variable, VIF quantifies how much the variance of the estimated regression coefficient is inflated due to collinearity with other predictors.

A high VIF (greater than 5 or 10) indicates that the predictor is highly correlated with other variables, suggesting multicollinearity.
