```python
# import necessary libraries
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import sklearn as scikit_learn

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('model_data.csv', index_col=0)
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.head(3)
```

```python
# Define the dependent Y and independent variables X
X = df[['Population']]
Y = df['Productivity']

# Add constant term to the independent variables
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.OLS(Y, X)
results = model.fit()

# Get the residuals
residuals = results.resid

# Print the least squares results
print(results.summary())

# Get the predicted values (fitted values)
y_pred = results.fittedvalues

# Get the confidence intervals for the predicted values
predictions = results.get_prediction(X)
conf_int = predictions.conf_int()

# Extract the lower and upper bounds of the confidence intervals
lower_bound = conf_int[:, 0]
upper_bound = conf_int[:, 1]

# Create a scatter plot of the data
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Population', y='Productivity', data=df, color='blue', label='Data')

# Plot the regression line
plt.plot(df['Population'], y_pred, color='red', label='OLS Regression Line')

# Plot the confidence intervals as shaded regions
plt.fill_between(df['Population'], lower_bound, upper_bound, color='blue', alpha=0.2, label='95% Confidence Interval')

# Add labels and title
plt.xlabel('Population')
plt.ylabel('Productivity')
plt.title('OLS Regression Line: Productivity vs Population')
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the dependent Y and independent variables X
X = df[['Population', 'Gdp', 'Compensation', 'CapitalStock', 'earlyLeavers']]
Y = df['Productivity']

# Add constant term to the independent variables
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(Y, X)
results = model.fit()

# Print the OLS regression results
print(results.summary())

# Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF results
print(vif_data)
```

VIF Calculation: For each independent variable, VIF quantifies how much the variance of the estimated regression coefficient is inflated due to collinearity with other predictors.

A high VIF (greater than 5 or 10) indicates that the predictor is highly correlated with other variables, suggesting multicollinearity.

```python
# Define the dependent Y and independent variables X
X = df[['Compensation', 'CapitalStock', 'earlyLeavers']]
Y = df['Productivity']

# Add constant term to the independent variables
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.OLS(Y, X)
results = model.fit()

# Print the OLS regression results
print(results.summary())

# Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF results
print(vif_data)
```
