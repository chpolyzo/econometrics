```python
# import necessary libraries
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy import stats
import sklearn as scikit_learn

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('model_data.csv', index_col=0)
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.head(3)
```

```python
# Define the dependent Y and independent variables X
X = df_normalized[['Population', 'Gdp', 'Compensation', 'CapitalStock']]
Y = df_normalized['Productivity']

# Add constant term to the independent variables
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(Y, X).fit()

# Print the logistic regression results
print(model.summary())

# Get residuals
residuals = model.resid
fitted = model.fittedvalues
```

```python
plt.scatter(fitted, Y, alpha=0.7)
plt.xlabel('Fitted Values (Predicted Productivity)')
plt.ylabel('Actual Productivity')
plt.title('Fitted vs Actual Values')
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--')  # 45-degree line
plt.show()
```

```python
fig = sm.graphics.plot_partregress_grid(model)
fig.tight_layout(pad=1.0)
plt.show()
```

```python
# 95% confidence intervals for model coefficients
conf_int = model.conf_int(alpha=0.05)
conf_int.columns = ['2.5%', '97.5%']
print(conf_int)
```

# Residuals vs Fitted (Linearity & Homoscedasticity)

```python
sns.residplot(x=fitted, y=residuals, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.axhline(0, color='black', linestyle='--')
plt.show()
```

# Normality of Residuals

```python
# plot the residuals histogram
residuals.hist(bins=20, edgecolor='black')
plt.title('Histogram of Residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()
```

```python
sm.qqplot(residuals, line='45', fit=True)
plt.title('Q-Q Plot of Residuals')
plt.show()
```

# Shapiro-Wilk Test (Statistical Test for Normality)

```python
shapiro_test = stats.shapiro(residuals)
print(f"Shapiro-Wilk p-value: {shapiro_test.pvalue:.4f}")
```

- p > 0.05: residuals likely normal.
- p < 0.05: residuals deviate from normality.
  
0.4823 > 0.5 hich sugeests normally distributed residuals
